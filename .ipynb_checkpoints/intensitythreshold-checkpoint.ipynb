```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Load and display image

im = cv2.imread('raheem.jpeg')
plt.imshow(im)
plt.axis('off')
plt.show()
```

```python
# Intensity-threshold histogram

vals = im.mean(axis=2).flatten()
# plot histogram with 255 bins
b, bins, patches = plt.hist(vals, 255)
plt.xlim([0,255])
plt.axvline(x=70, linestyle='dashed')
plt.axvline(x=138, linestyle='dashed')
plt.savefig("intensity.png")
```

```python
# Display post-intensity threshold segmented image

x = cv2.threshold(im, 70, 138, cv2.THRESH_BINARY)
plt.imshow(x[1])
plt.axis('off')
plt.savefig('sterlingseg.png')
```
